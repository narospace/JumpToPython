````python
class Calculator:
    def __init__(self):
        self.value = 0

    def add(self, val):
        self.value += val
        
# ë¶€ëª¨(ê¸°ë°˜) í´ë˜ìŠ¤ì˜ ê¸°ëŠ¥ í™•ì¥
class UpgradeCalculator(Calculator):
    def minus(self, val):
        self.value -= val

cal = UpgradeCalculator()

cal.add(10)
print("add(10) :", cal.value)

cal.minus(7)
print("minus(7):", cal.value)
````
