```python
# *args : <class 'tuple'> 반복객체 튜플 -> 즉 여러개의 객체(배열 객체)를 요소로 갖는 자료형이 필요
def get_avg(*args):
    print(type(args)) # args = (1, 2, 3, 4, 5)
    total = 0
    for i in args:
        total += i
        print(i, total)
    return total / len(args)

print("평균:", get_avg(1, 2, 3, 4, 5))
```

Q5
다음은 "test.txt"라는 파일에 "Life is too short" 문자열을 저장한 후 다시 그 파일을 읽어서 출력하는 프로그램이다.

f1 = open("test.txt", 'w')
f1.write("Life is too short")

f2 = open("test.txt", 'r')
print(f2.read())
이 프로그램은 우리가 예상한 "Life is too short"라는 문장을 출력하지 않는다. 우리가 예상한 값을 출력할 수 있도록 프로그램을 수정해 보자.

```python
print("프로그램 시작=================")

try:
    f1 = open("test.txt", 'w')
    f1.write("Life is too short")
except Exception as ex:
    print(ex)
finally:
    f1.close()
print("f1.closed:",f1.closed)

try:
    f2 = open("test.txt", 'r')
    print(f2.read())
except Exception as ex:
    print(ex)
finally:
    f2.close()
print("f2.closed:",f2.closed)

print("프로그램 종료=================")
```

Q6
사용자의 입력을 파일(test.txt)에 저장하는 프로그램을 작성해 보자. (단 프로그램을 다시 실행하더라도 기존에 작성한 내용을 유지하고 새로 입력한 내용을 추가해야 한다.)

```python
data_in = input("메시지를 입력해 주세요: ")

# 기존 파일에 추가 메시지 저장
with open("test.txt", 'a', encoding='utf-8') as fw:
    fw.write('\n' + data_in)

with open("test.txt", 'r', encoding='utf-8') as fo:
    print(fo.read())
```

Q7
다음과 같은 내용을 지닌 파일 test.txt가 있다. 이 파일의 내용 중 "java"라는 문자열을 "python"으로 바꾸어서 저장해 보자.

Life is too short
you need java
※ replace 함수를 사용해 보자.

```python
# test.txt 읽어와 문자열.replace("java", "python") 처리 후 파일 저장

with open("test.txt", 'rt', encoding='utf-8') as fr:
    data = fr.read().replace("java", "python")

with open("test.txt", 'wt', encoding='utf-8') as fw:
    fw.write(data)
```
